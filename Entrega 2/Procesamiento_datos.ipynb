```python
import pandas as pd
```

```python
df = pd.read_csv('UCI_Credit_Card.csv')
```

```python
df
```

```python
def gender(sex):
    if sex == 1:
        return 'Gen - Hombre'
    elif sex == 2:
        return 'Gen - Mujer'
    else:
        return 'Gen - Sin Información'

def education(ed):
    if ed == 1:
        return 'Ed - Posgrado'
    elif ed == 2:
        return 'Ed - Universitario'
    elif ed == 3:
        return 'Ed - Bachillerato'
    elif ed == 4:
        return 'Ed - Otros'
    else:
        return 'Ed - Sin Información'
    
def marriage(status):
    if status == 1:
        return 'Mat - Casado'
    elif status == 2:
        return 'Mat - Soltero'
    elif status == 3:
        return 'Mat - Otros'
    else:
        return 'Mat - Sin Información'
```

***De-codificamos las variables***

```python
df['COD_SEX'] = df.SEX.apply(gender)
df['COD_EDUCATION'] = df.EDUCATION.apply(education)
df['COD_MARRIAGE'] = df.MARRIAGE.apply(marriage)
```

***Hacemos el encoding para que las categorias queden como numeros***

```python
df_gen = pd.get_dummies(df.COD_SEX)['Gen - Mujer']
```

```python
df_ed = pd.get_dummies(df.COD_EDUCATION).drop('Ed - Sin Información', axis = 1)
```

```python
df_mat = pd.get_dummies(df.COD_MARRIAGE).drop('Mat - Sin Información', axis = 1)
```

***DataFrame FInal***

```python
pd.options.display.max_columns = None
```

```python
df_fin = df[['LIMIT_BAL', 'BILL_AMT1','BILL_AMT2','BILL_AMT3','BILL_AMT4','BILL_AMT5','BILL_AMT6',
             'PAY_AMT1','PAY_AMT2','PAY_AMT3','PAY_AMT4','PAY_AMT5','PAY_AMT6',
             'PAY_0','PAY_2' ,'PAY_3','PAY_4','PAY_5','PAY_6'
            ]].copy()
```

```python
df_demo = pd.concat([df['AGE'], df_gen, df_ed, df_mat], axis = 1)
```

```python
df_final = pd.concat([df_fin, df_demo, df['default.payment.next.month']], axis = 1)
```
